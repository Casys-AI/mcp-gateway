# Security & Resource Limits

Demonstrates timeout, memory limits, and permission controls.

```typescript
import { DenoSandboxExecutor } from "jsr:@casys/mcp-gateway";

const sandbox = new DenoSandboxExecutor({ timeout: 2000, memoryLimit: 64 });
```

## Timeout Protection

```typescript
const result = await sandbox.execute("while(true) {}");
console.log(JSON.stringify(result, null, 2));
```

## File Access Blocked

```typescript
const result = await sandbox.execute(`
try {
  await Deno.readTextFile('/etc/passwd');
  return { blocked: false };
} catch (e) {
  return { blocked: true, error: e.message };
}
`);
console.log(JSON.stringify(result, null, 2));
```

## Network Access Blocked

```typescript
const result = await sandbox.execute(`
try {
  await fetch('https://example.com');
  return { blocked: false };
} catch (e) {
  return { blocked: true, error: e.message };
}
`);
console.log(JSON.stringify(result, null, 2));
```
